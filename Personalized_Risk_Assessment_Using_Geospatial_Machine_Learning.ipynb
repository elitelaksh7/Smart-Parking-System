```python
import folium
import pandas as pd
from folium.plugins import HeatMap
```

```python
data = {
    'Region ID': [
        'Region A', 'Region B', 'Region C', 'Region D', 'Region E', 'Region F', 'Region G', 'Region H', 'Region I', 'Region J',
        'Region K', 'Region L', 'Region M', 'Region N', 'Region O', 'Region P', 'Region Q', 'Region R', 'Region S', 'Region T'
    ],
    'State': [
        'California', 'California', 'California', 'California', 'California', 'California',
        'Texas', 'Texas', 'Texas', 'Texas',
        'New York', 'New York', 'New York', 'New York', 'New York', 'New York',
        'Florida', 'Florida', 'Florida', 'Florida'
    ],
    'Avg Vulnerability Score': [
        75, 80, 60, 50, 90, 85, 30, 60, 40, 50,
        60, 55, 50, 65, 70, 80, 45, 60, 50, 55
    ],
    'Latitude': [
        36.7783, 36.7785, 36.7779, 36.7774, 36.7768, 36.7771, 31.9686, 31.9683, 31.9681, 31.9684,
        40.7128, 40.7132, 40.7135, 40.7138, 40.7141, 40.7144, 27.9944, 27.9950, 27.9960, 27.9971
    ],
    'Longitude': [
        -119.4179, -119.4180, -119.4175, -119.4173, -119.4176, -119.4174, -99.9018, -99.9021, -99.9022, -99.9020,
        -74.0060, -74.0057, -74.0055, -74.0053, -74.0050, -74.0048, -81.7601, -81.7604, -81.7606, -81.7607
    ]
}
```

```python
df = pd.DataFrame(data)
```

```python
states_of_interest = ['California', 'Texas', 'New York', 'Florida']
df_filtered = df[df['State'].isin(states_of_interest)]
```

```python
m = folium.Map(location=[37.8, -96], zoom_start=5)
```

```python
heat_data = [[row['Latitude'], row['Longitude'], row['Avg Vulnerability Score']] for index, row in df_filtered.iterrows()]
```

```python
HeatMap(heat_data, radius=25, blur=15).add_to(m)
```

```python
def get_color(vulnerability_score):
    if vulnerability_score < 50:
        return 'green'
    elif vulnerability_score < 70:
        return 'orange'
    else:
        return 'red'
```

```python
for idx, row in df_filtered.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=folium.Popup(f"Region: {row['Region ID']}<br>State: {row['State']}<br>Vulnerability Score: {row['Avg Vulnerability Score']}<br>", parse_html=True),
        icon=folium.Icon(color=get_color(row['Avg Vulnerability Score']))
    ).add_to(m)
```
